```python
#Load the required libraries

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (accuracy_score,precision_score,recall_score, confusion_matrix,ConfusionMatrixDisplay,f1_score,classification_report,)
```

```python
#Load the data
obesity = pd.read_csv('ObesityDataSet.csv')
obesity.head()
```

```python
#Summary of the dataset
obesity.info()
```

```python
#Address categorical data features with one hot encoding
pre_obesity = pd.get_dummies(obesity,columns=['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'],drop_first=False)
pre_obesity.head()
```

```python
#Separate the independent and dependent variables
X = pre_obesity.drop('NObesity', axis=1)
y = pre_obesity['NObesity']
```

```python
#Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=100)
```

```python
#Use the Gaussian Naive Bayes algorithm to train the model
model = GaussianNB()

model.fit(X_train, y_train);
```

```python
#Check the evaluation metrics of the model
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_pred, y_test)
precision = precision_score(y_test, y_pred, pos_label='positive', average='macro')
recall = recall_score(y_test, y_pred, pos_label='positive', average='macro')
print("Accuracy:", accuracy*100)
print("Precision:", precision*100)
print("Recall:", recall*100)
```
